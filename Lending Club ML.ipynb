```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# might be needed depending on your version of Jupyter
%matplotlib inline

df = pd.read_csv('DATA/lending_club_loan_two.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
sns.countplot(x=df.loan_status)
```

```python
plt.figure(figsize=(11,5))
sns.histplot(x=df.loan_amnt, bins=30)
```

```python
plt.figure(figsize=(12,7))
sns.heatmap(df.corr(), annot=True, cmap='viridis')
```

```python
sub_grade_sorted = sorted(df['sub_grade'].unique())
plt.figure(figsize=(12,4))
sns.countplot(data=df, x='sub_grade', order=sub_grade_sorted, palette='coolwarm')
```

```python
plt.figure(figsize=(12,4))
sns.countplot(data=df, x='sub_grade', order=sub_grade_sorted, hue='loan_status', palette='coolwarm')
```
